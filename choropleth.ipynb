```python
import pandas as pd
import folium
import sqlite3

conn = sqlite3.connect('test.db')
c = conn.cursor()

locdata = pd.read_sql_query('select l.neighborhood as Neighborhood, Count(l.neighborhood) as Number from locations l GROUP BY neighborhood;', 
                           conn)
print(locdata.head(8))

m = folium.Map(location = [41.795, -87.585])
# Depreciated, needs updating
m.choropleth(geo_data='map\\ucpd-boundary.geojson',
            data = locdata, 
            columns = ['Neighborhood', 'Number'],
            key_on = 'feature.properties.Name',
            fill_color = 'YlOrRd',
            highlight = True)

display(m)
```
